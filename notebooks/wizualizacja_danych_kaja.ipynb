```python
import pandas as pd
kanalzero=pd.read_csv("../data/processed/processed_KanalZero_videos_info.csv")
orb=pd.read_csv("../data/processed/processed_ORB_videos_info.csv")
konop=pd.read_csv("../data/processed/processed_Konopskyy_videos_info.csv")
```

```python

kanalzero["published_at"] = pd.to_datetime(kanalzero["published_at"],errors="coerce")
kanalzero = kanalzero.sort_values("published_at")
kanalzero=kanalzero[kanalzero["published_at"]>"2024-01-01"]
kanalzero.head()
```

```python
orb["published_at"] = pd.to_datetime(orb["published_at"],errors="coerce")
orb = orb.sort_values("published_at")
orb.head()
```

```python
konop["published_at"] = pd.to_datetime(konop["published_at"],errors="coerce")
konop = konop.sort_values("published_at")
konop=konop[konop["published_at"]>"2021-01-01"]
konop.head()
```
